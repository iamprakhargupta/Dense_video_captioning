```python
from glob import glob
```

```python
folders=glob("E:\\youcook_code\\data\\raw_videos\\raw_videos\\validation\\*")
```

```python
outpath="E:\\youcook_code\\data\\raw_videos\\raw_videos\\val_asr"
import os
```

```python
len(folders)
```

```python
for folder in folders:
    names=folder.split('\\')[-1]
    asr_out=os.path.join(outpath, names)
    try:
        os.mkdir(asr_out)
    except OSError as error:
        print(error)    
    paths=folder+"\\*"
    files=glob(paths)
    for file in files:
        text = model.transcribe(file)
        videoname=file.split("\\")[-1]
        videoname=videoname.split(".")[0]+".txt"
        outfile=os.path.join(asr_out,videoname)
        print(outfile)
        with open(outfile,"w+") as f:
            f.write(text['text'])
        
```
